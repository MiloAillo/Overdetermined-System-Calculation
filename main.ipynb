### How to Solve Overdetermined System of Linear Equation in Python using NumPy

we're gonna use Normal Equation formula to find the best-fit solution of the overdetermined system.

Now let us install and load NumPy to our project.

```python
%pip install numpy

import numpy as np
```

Now we need to have Coefficient Matrix and Constants Vector according to the Ax = b formula.
But no unknown variables here because the output is the unknown variable itself so i dont have to initialize anything here.

```python
A = np.array([
    [8, 7],
    [1, 3],
    [4, 1]
])

b = np.array([
    [2],
    [7],
    [-15]
])
```

lets find the transposed of the Coefficient Matrix,

```python
transposed_A = A.transpose()

print(transposed_A)
```

lets multiply the the transposed A to A, like what the left-hand side of the formula show.

```python
approximation_coefficient_matrix = np.matmul(transposed_A, A)

print(approximation_coefficient_matrix)
```

now do the same with the right-hand side, tranposed A to b.

```python
approximation_constants_vector = np.matmul(transposed_A, b)

print(approximation_constants_vector)
```

now we have the normal system of linear equation that holds the approximation solution to the previous overdetermined system.

to find the solution, we have to look at the formula,
#### Ax = b
we need to find the solution (x), so we need to remove the A.
normal algebraic form involve dividing both LHS and RHS value, but matrix is different. 
To remove a matrix, we need to multiply the matrix to the its inverse self.

but because this is a code, we can just ignore them and put the same inversed approximation coefficient matrix on the RHS, and that mean multiplying the approximation constants vector with it.

```python
inversed_approximation_coefficient_matrix = np.linalg.inv(approximation_coefficient_matrix)

best_fit_solution = np.matmul(inversed_approximation_coefficient_matrix, approximation_constants_vector)

print(best_fit_solution)
```

yay we got our best-fit solution for our overdetermined system. Lets look at the graph to see if its correct,
